```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# Set the number of days to simulate
N = 365  # One year of data

# Generate random data: Number of customers per day (between 0 and 10)
np.random.seed(42)  # For reproducibility
data = np.random.randint(0, 11, size=N)

# Compute PMF (Probability Mass Function)
counts = Counter(data)
total_days = sum(counts.values())
pmf = np.array([counts[i] / total_days for i in range(11)])

# Compute CDF (Cumulative Distribution Function)
cdf = np.cumsum(pmf)

# Unique values (0 to 10 customers)
unique = np.arange(11)
# PMF Plot
plt.figure(figsize=(10, 5))
sns.barplot(x=unique, y=pmf, color="skyblue", edgecolor="black")
plt.xlabel("Number of Customers")
plt.ylabel("Probability Mass Function (PMF)")
plt.title("Probability Mass Function of Customer Visits")
plt.xticks(unique)

plt.show()

# CDF Plot
plt.figure(figsize=(10, 5))
sns.lineplot(x=unique, y=cdf, marker='o', color="red", linestyle='-')
plt.xlabel("Number of Customers")
plt.ylabel("Cumulative Distribution Function (CDF)")
plt.title("Cumulative Distribution Function of Customer Visits")
plt.xticks(unique)
plt.ylim(0, 1)  # Set y-axis range
plt.grid()
plt.show()
```

```python


import scipy.stats as stats

# Given data
N = 100  # Number of students
p = 0.7  # Probability of passing

# i) Probability that exactly 60 students pass
prob_60 = stats.binom.pmf(60, N, p)

# ii) Probability that at least 75 students pass (P(X ≥ 75) = 1 - P(X < 75) = 1 - P(X ≤ 74))
prob_75_or_more = 1 - stats.binom.cdf(74, N, p)

# iii) Expected number of students passing
expected_students = N * p

# Print results
print(f"Probability that exactly 60 students pass: {prob_60:.4f}")
print(f"Probability that at least 75 students pass: {prob_75_or_more:.4f}")
print(f"Expected number of students passing: {expected_students}")
```
